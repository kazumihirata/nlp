12． 1列目をcol1.txtに，2列目をcol2.txtに保存  
各行の1列目だけを抜き出したものをcol1.txtに，2列目だけを抜き出したものをcol2.txtとしてファイルに保存せよ．確認にはcutコマンドを用いよ．



```
!ls
!rm col*.txt
!ls
```

```
with open('hightemp.txt') as f:
  lines = f.readlines()

for num, line in enumerate(lines) :
  first = line.split("\t")[0]
  second = line.split("\t")[1]
  with open('col1.txt','a') as f:
    f.write(first+"\n")
  with open('col2.txt','a') as f:
    f.write(second+"\n")
```

13． col1.txtとcol2.txtをマージ  
12で作ったcol1.txtとcol2.txtを結合し，元のファイルの1列目と2列目をタブ区切りで並べたテキストファイルを作成せよ．確認にはpasteコマンドを用いよ．



```
f = open('col1.txt')
cols1 = f.readlines()
f.close()
# col1
f = open('col2.txt')
cols2 = f.readlines()
f.close()
# col2
for (col1, col2) in zip(cols1, cols2) :
  col1 = col1.replace("\n", "")
  col2 = col2.replace("\n", "")
  with open('concat_col1_col2.txt','a') as f:
    f.write(col1 + "\t" + col2 + "\n")
```

14． 先頭からN行を出力  
自然数Nをコマンドライン引数などの手段で受け取り，入力のうち先頭のN行だけを表示せよ．確認にはheadコマンドを用いよ．



```
import random

with open('hightemp.txt') as f:
  lines = f.readlines()

input_num = random.randrange(1, len(lines)+1)
print(input_num) # N
for num, line in enumerate(lines) :
  if input_num <= num :
    break
  print(line.replace("\n", ""))

print(num) # loop数
```

15． 末尾のN行を出力

```
import random

with open('hightemp.txt') as f:
  lines = f.readlines()

input_num = random.randrange(1, len(lines)+1)
print(input_num) # N

from_line = len(lines) - input_num

for num, line in enumerate(lines) :
  if from_line <= num :
    print(line.replace("\n", ""))
```

16． ファイルをN分割する  
自然数Nをコマンドライン引数などの手段で受け取り，入力のファイルを行単位でN分割せよ．同様の処理をsplitコマンドで実現せよ．



```
import random

def line_counts(max_count, n):
  # 商を求める
  quo = max_count // n
  # 余りを求める
  rem = max_count % n
  return [quo+1] * rem + [quo] * (n - rem)


with open('hightemp.txt') as f:
  lines = f.readlines()

input_num = random.randrange(1, len(lines)+1)
# input_num = 1 # 24
# input_num = 2 # 12, 24
# input_num = 3 # 8, 16, 24
# input_num = 4 # 6, 12, 18, 24
# input_num = 5 # 5, 10, 15, 20, 24
# input_num = 6 # 4, 8, 12, 16, 20, 24
# input_num = 7 # 3, 6, 9, 12, 15, 18, 21, 24 ....?
# input_num = 8 # 3, 6, 9, 12, 15, 18, 21, 24
# input_num = 9
# input_num = 10
# input_num = 11
print(input_num)

# N分割するための組み合わせを算出
rows_list = []
rows_list = line_counts(len(lines), input_num)
print(rows_list)

from_num = 0
to_num = 0
for num, row in enumerate(rows_list) :
  file_name = 'split{}.txt'.format(num+1)
  from_num = num * row
  to_num = (num + 1) * row
  for i in lines[from_num:to_num] :
    with open(file_name,'a') as f:
      f.write(i)
```

19． 各行の1コラム目の文字列の出現頻度を求め，出現頻度の高い順に並べる
各行の1列目の文字列の出現頻度を求め，その高い順に並べて表示せよ．確認にはcut, uniq, sortコマンドを用いよ．

```
with open('hightemp.txt') as f:
    lines = f.readlines()
ones = []

for line in lines :
  ones.append(line.split("\t")[0])
print(ones)

counter = {}

for one in ones :
  if not one in counter :
    counter[one] = 1
  else :
    tmp_count = counter[one]
    counter[one] = tmp_count + 1
print(counter)

for k, v in sorted(counter.items(), key=lambda x: x[1], reverse=True):
    print(k)
```
