04． 元素記号  
"Hi He Lied Because Boron Could Not Oxidize Fluorine. New Nations Might Also Sign Peace Security Clause. Arthur King Can."という文を単語に分解し，1, 5, 6, 7, 8, 9, 15, 16, 19番目の単語は先頭の1文字，それ以外の単語は先頭に2文字を取り出し，取り出した文字列から単語の位置（先頭から何番目の単語か）への連想配列（辞書型もしくはマップ型）を作成せよ．

```
target = 'Hi He Lied Because Boron Could Not Oxidize Fluorine. New Nations Might Also Sign Peace Security Clause. Arthur King Can.'
target = target.replace('.', '')
word_dict = {}

list = [1, 5, 6, 7, 8, 9, 15, 16, 19]

for num, i in enumerate(target.split(' ')) :
  num += 1
  if num in list :
    word_dict[num] = i[:1]
  else :
    word_dict[num] = i[:2]
  
word_dict
```

05． n-gram  
与えられたシーケンス（文字列やリストなど）からn-gramを作る関数を作成せよ．この関数を用い，"I am an NLPer"という文から単語bi-gram，文字bi-gramを得よ．

```
# 単語bi-gram -> ['I am', 'am an', an NLPer']
# 文字bi-gram -> ['Ia', 'am', 'ma', 'an', 'nN', 'NL', 'LP', 'Pe', 'er']
target = 'I am an NLPer'

def ngram(input, n):
    last = len(input) - n + 1
    ret = []
    for i in range(0, last):
        ret.append(input[i:i+n])
    return ret
  
tango_bigram = ngram(target.replace(' ', ''), 2)
tango_bigram
```

```
moji_bigram = ngram(target.split(' '), 2)
moji_bigram
```

06． 集合  
"paraparaparadise"と"paragraph"に含まれる文字bi-gramの集合を，それぞれ, XとYとして求め，XとYの和集合，積集合，差集合を求めよ．さらに，'se'というbi-gramがXおよびYに含まれるかどうかを調べよ．

```
def ngram(input, n):
    last = len(input) - n + 1
    ret = []
    for i in range(0, last):
        ret.append(input[i:i+n])
    return ret

def remove_duplicate(list) :
  results = []
  for i in range(len(list)) :
    if not list[i] in results :
      results.append(list[i])
  return results

def union_set(list1, list2) :
  results = remove_duplicate(list1)
  
  for i in range(len(list2)) :
    if not list2[i] in results :
      results.append(list2[i])
  return results

def intersection_set(list1, list2) :
  results = []
  
  for i in range(len(list1)) :
    if list1[i] in list2 :
      results.append(list1[i])
  return results

def difference_set(list1, list2) :
  results = []
  for i in range(len(list1)) :
    if not list1[i] in list2 :
      results.append(list1[i])
  return results


X = ngram("paraparaparadise", 2)
Y = ngram("paragraph", 2)

remove_dups_X = remove_duplicate(X)
remove_dups_Y = remove_duplicate(Y)

wa = union_set(remove_dups_X, remove_dups_Y)
seki = intersection_set(remove_dups_X, remove_dups_Y)
sa = difference_set(remove_dups_X, remove_dups_Y)
```

08． 暗号文  
与えられた文字列の各文字を，以下の仕様で変換する関数cipherを実装せよ．  

英小文字ならば(219 - 文字コード)の文字に置換  
その他の文字はそのまま出力  
この関数を用い，英語のメッセージを暗号化・復号化せよ．



```
d = {
  'a': 97,
  'b': 98,
  'c': 99,
  'd': 100,
  'e': 101,
  'f': 102,
  'g': 103,
  'h': 104,
  'i': 105,
  'j': 106,
  'k': 107,
  'l': 108,
  'm': 109,
  'n': 110,
  'o': 111,
  'p': 112,
  'q': 113,
  'r': 114,
  's': 115,
  't': 116,
  'u': 117,
  'v': 118,
  'w': 119,
  'x': 120,
  'y': 121,
  'z': 122,
}

def cipher(target) :
  result = ''
  for i in target :
    tmp = i
    if i.islower() :
      tmp = str(219 - d[i])
    result += tmp
  return result
```

09． Typoglycemia  
スペースで区切られた単語列に対して，各単語の先頭と末尾の文字は残し，それ以外の文字の順序をランダムに並び替えるプログラムを作成せよ．  
ただし，長さが４以下の単語は並び替えないこととする．  
適当な英語の文（例えば"I couldn't believe that I could actually understand what I was reading : the phenomenal power of the human mind ."）を与え，その実行結果を確認せよ．

```
import random

target = "I couldn't believe that I could actually understand what I was reading : the phenomenal power of the human mind ."

target = target.replace(',', '')
target = target.replace('.', '')
target = target.replace(':', '')
target = target.replace('  ', ' ')
# trim
target = target.strip()
list = target.split(' ')

results = []
for word in list :
  if len(word) <= 4 :
    results.append(word)
  else :
    results.append(word[0] + word[-1:])


print(results)
random.shuffle(results)
print(results)
```
